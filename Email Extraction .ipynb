```python
modified_links = []
for i in range(len(links)):
    link_a = links[i] + 'overlay/contact-info/'
    modified_links.append(link_a)
modified_links
len(modified_links)
```

```python
contact_info_x = []
for link in tqdm(modified_links):
    driver.get(link)
    contact_info_basic = driver.find_elements(By.CLASS_NAME, 'pv-contact-info')
    contact_info_structured = contact_info_basic[0].text.split('\n')[1:]
    contact_info_x.append(contact_info_structured)
    
len(contact_info_x)
```

```python
contact_info = []
for conn_contact in contact_info_x:
    contact_dict = {}
    for idx in range(0, len(conn_contact)-2, 2):
        contact_dict[conn_contact[idx]] = conn_contact[idx + 1]
    contact_info.append(contact_dict)
len(contact_info)
```
